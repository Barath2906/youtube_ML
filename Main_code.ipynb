```python
import pandas as pd

df = pd.DataFrame(columns=['video_id', 'VideoNum', 'tagNo', 'tagName'])

tag_dict = {}

video_id_dict = {}
video_id_start = 100

tag_counter = 1

def append_row(df, row):
    return pd.concat([
                df, 
                pd.DataFrame([row], columns=row.index)]
           ).reset_index(drop=True)


for index, row in dfCluster.iterrows():
    video_id = row['video_id']
    
    if video_id not in video_id_dict:
        video_id_dict[video_id] = video_id_start
        video_id_start += 1

    tags = [tag.strip() for tag in row['video_tags'].split(',')]
    
    for tag in tags:
        if tag not in tag_dict:
            tag_dict[tag] = tag_counter
            tag_counter += 1

        new_row = pd.Series({
            'video_id': video_id, 
            'VideoNum': video_id_dict[video_id], 
            'tagNo': tag_dict[tag],
            'tagName': tag
        })
        df = append_row(df, new_row)

print(df)
```

```python
import pandas as pd
from sklearn.cluster import KMeans


X = df[['VideoNum', 'tagNo']]

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

def recommend_videos_by_tag(tag_name, df, num_recommendations=20):
    if tag_name not in df['tagName'].values:
        print(f"Tag name '{tag_name}' not found in the dataset.")
        return pd.DataFrame(columns=['video_id', 'tagName'])
    
    tag_cluster = df.loc[df['tagName'] == tag_name, 'Cluster'].values[0]
    
    cluster_videos = df[(df['Cluster'] == tag_cluster) & (df['tagName'] != tag_name)]
    
    recommendations = cluster_videos.head(num_recommendations)
    return recommendations[['video_id', 'tagName']]

tag_to_recommend = 'vindooly skits'
recommended_videos = recommend_videos_by_tag(tag_to_recommend, df)
print(f"Recommended videos for tag '{tag_to_recommend}':")
print(recommended_videos)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the dataset (replace with your path)
video_data = pd.read_csv("C:/Users/BARATH KANNAN/OneDrive/Desktop/final_project/data.csv")

video_data['tagName'] = video_data['tagName'].fillna('')

vectorizer = TfidfVectorizer(stop_words='english')

X = vectorizer.fit_transform(video_data['tagName'])

num_clusters = 150

kmeans = KMeans(n_clusters=num_clusters, random_state=42)

video_data['cluster'] = kmeans.fit_predict(X)

silhouette_avg = silhouette_score(X, video_data['cluster'])
print(f'Silhouette Score for {num_clusters} clusters: {silhouette_avg}')

print(video_data[['video_title', 'tagName', 'cluster']].head())
```
